```python
from costometer.utils.plotting_utils import set_font_sizes
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
set_font_sizes()
```

# Example Simulated Data

```python
simulation_data = pd.read_csv("data/processed/simulated/high_increasing/MCL/linear_depth/search_space/1729_habitual_baseline_null.csv", index_col=False)
```

```python
simulation_data.head
```

```python
simulation_data.groupby(["pid", "i_episode", "sim_cost_parameter_values"], as_index=False).mean()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.lineplot(y="return", x="i_episode", hue="sim_cost_parameter_values",data=simulation_data.groupby(["pid", "i_episode", "sim_cost_parameter_values"], as_index=False).mean())
```

```python
simulation_data.groupby(["pid", "i_episode", "sim_cost_parameter_values"], as_index=False).sum()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.lineplot(y="num_early", x="i_episode", hue="sim_cost_parameter_values",data=simulation_data.groupby(["pid", "i_episode", "sim_cost_parameter_values"], as_index=False).sum())
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.lineplot(y="unbounded_present_bias", x="i_episode", hue="sim_cost_parameter_values",data=simulation_data.groupby(["pid", "i_episode", "sim_cost_parameter_values"], as_index=False).mean())
```

# Human Data

```python
actual_human_data = pd.read_csv("../../../data/processed/methods_main/mouselab-mdp.csv")
actual_human_data.head
```

```python
list(actual_human_data)
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.lineplot(y="score", x="trial_index", data= actual_human_data)
plt.xlabel("Trial number")
plt.xlabel("Average reward per trial")
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.lineplot(y="num_early", x="trial_index", data= actual_human_data)
plt.xlabel("Trial number")
plt.xlabel("Number of early nodes clicked per trial")
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.lineplot(y="num_late", x="trial_index", data= actual_human_data)
plt.xlabel("Trial number")
plt.xlabel("Number of late nodes clicked per trial")
```
