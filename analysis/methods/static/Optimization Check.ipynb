```python
full_data = {}
for pid in pids:
    matching_files = list(Path("../../..").glob(f"cluster/data/logliks_opt/back_dist_depth_eff_forw2/methods_main/*_{pid}_*"))
    
    for curr_file_idx, curr_file in enumerate(matching_files):
        with open(curr_file, "rb") as f:
            curr_data = pickle.load(f)
        
        if curr_file_idx == 0:
            participant_data = {key : val for key, val in curr_data.items() if key != "Null"}
        else:
            for model_name, model_data in curr_data.items():
                if model_name != "Null":
                    if model_data["trials"].best_trial["result"]["loss"] < \
                    participant_data[model_name]["trials"].best_trial["result"]["loss"]:
                        participant_data[model_name] = model_data
                        
    if len(matching_files)>0:
        full_data[pid] = participant_data
```

```python
map_df = []
for pid, participant_data in full_data.items():
    for model_name, model_data in participant_data.items():
        num_actions = sum([len(trial) for trial in model_data["res"][0][-20:]])
        print(model_name, model_data["trials"].best_trial["misc"]["vals"])
        map_df.append([model_name, model_data["trials"].best_trial["result"]["loss"], pid, num_actions])#, len(model_data["trials"].best_trial["misc"]["vals"])])
```

```python
map_df = pd.DataFrame(map_df, columns = ["Model Name", "MAP", "PID", "Num Actions"])#, "Num Parameters"])
```

```python
with open("parameter_numbers.pickle", "rb") as f:
    parameter_map = pickle.load(f)
```

```python
map_df["Num Parameters"] = map_df["Model Name"].apply(lambda val : parameter_map[val])
```

```python
sum_test = map_df.groupby(["Model Name","Num Parameters"]).sum().reset_index()
```

```python
from statsmodels.tools.eval_measures import bic

sum_test["bic"] = sum_test.apply(
    lambda row: bic(
    llf=-row["MAP"],
    nobs=row["Num Actions"],
    df_modelwc=row["Num Parameters"], # temp, discount rate, alpha, forw, eff, dist\n",
    ),
    axis=1)
```
